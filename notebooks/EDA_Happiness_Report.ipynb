Este notebook contiene un análisis exploratorio de datos (EDA) de los informes de felicidad mundial de 2015 a 2019.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

<a id='2015'></a>
# EDA - Dataset 2015

```python
# Cargar datos 2015
df_2015 = pd.read_csv('csv/2015.csv')
print(f"\nDimensiones: {df_2015.shape}")
df_2015.head()
```

```python
# Información del dataset
print(df_2015.info())
```

```python
# Estadísticas descriptivas 2015
df_2015.describe()
```

```python
# Verificar valores nulos
print(df_2015.isnull().sum())
```

```python
# Top 10 países más felices en 2015
print(df_2015.head(10))
```

```python
# Visualizaciones 2015
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Análisis Visual - Dataset 2015', fontsize=16, fontweight='bold')

# Histograma del Happiness Score
axes[0, 0].hist(df_2015.iloc[:, 3], bins=30, color='skyblue', edgecolor='black')
axes[0, 0].set_title('Distribución del Happiness Score 2015')
axes[0, 0].set_xlabel('Happiness Score')
axes[0, 0].set_ylabel('Frecuencia')

# Top 10 países
top10_2015 = df_2015.head(10)
axes[0, 1].barh(top10_2015.iloc[:, 0], top10_2015.iloc[:, 3], color='coral')
axes[0, 1].set_title('Top 10 Países Más Felices - 2015')
axes[0, 1].set_xlabel('Happiness Score')
axes[0, 1].invert_yaxis()

# Boxplot de variables numéricas
numeric_cols = df_2015.select_dtypes(include=[np.number]).columns[2:]
df_2015[numeric_cols].boxplot(ax=axes[1, 0], vert=False)
axes[1, 0].set_title('Boxplot de Variables Numéricas 2015')

# Matriz de correlación
corr_matrix = df_2015.select_dtypes(include=[np.number]).corr()
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', ax=axes[1, 1], cbar_kws={'shrink': 0.8})
axes[1, 1].set_title('Matriz de Correlación 2015')

plt.tight_layout()
plt.show()
```

```python
# Análisis Bivariado 2015

# Seleccionar columnas numéricas (excluyendo rank)
numeric_cols = df_2015.select_dtypes(include=[np.number]).columns[2:]
score_col = df_2015.select_dtypes(include=[np.number]).columns[2]

fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('Análisis Bivariado - Happiness Score vs Otras Variables (2015)', fontsize=16, fontweight='bold')
axes = axes.ravel()

for idx, col in enumerate(numeric_cols[1:]):
    if idx < 6:
        axes[idx].scatter(df_2015[col], df_2015[score_col], alpha=0.6, color='steelblue')
        axes[idx].set_xlabel(col)
        axes[idx].set_ylabel('Happiness Score')
        axes[idx].set_title(f'{col} vs Happiness Score')
        
        # Agregar línea de tendencia
        z = np.polyfit(df_2015[col].dropna(), df_2015[score_col][df_2015[col].notna()], 1)
        p = np.poly1d(z)
        axes[idx].plot(df_2015[col], p(df_2015[col]), "r--", alpha=0.8, linewidth=2)

plt.tight_layout()
plt.show()
```

# 🔄 Análisis Comparativo entre Años (2015-2019)

```python
# Comparación de estadísticas básicas
print("\nComparación de Número de Países por Año:")
years_count = {
    '2015': len(df_2015),
    '2016': len(df_2016),
    '2017': len(df_2017),
    '2018': len(df_2018),
    '2019': len(df_2019)
}
for year, count in years_count.items():
    print(f"{year}: {count} países")
```

```python
# Visualización comparativa de distribuciones
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('Distribución del Happiness Score por Año (2015-2019)', fontsize=16, fontweight='bold')
axes = axes.ravel()

# 2015
axes[0].hist(df_2015.iloc[:, 3], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
axes[0].set_title('2015')
axes[0].set_xlabel('Happiness Score')
axes[0].set_ylabel('Frecuencia')

# 2016
axes[1].hist(df_2016.iloc[:, 3], bins=20, color='lightgreen', edgecolor='black', alpha=0.7)
axes[1].set_title('2016')
axes[1].set_xlabel('Happiness Score')

# 2017
axes[2].hist(df_2017.iloc[:, 2], bins=20, color='lightcoral', edgecolor='black', alpha=0.7)
axes[2].set_title('2017')
axes[2].set_xlabel('Happiness Score')

# 2018
axes[3].hist(df_2018.iloc[:, 2], bins=20, color='gold', edgecolor='black', alpha=0.7)
axes[3].set_title('2018')
axes[3].set_xlabel('Happiness Score')
axes[3].set_ylabel('Frecuencia')

# 2019
axes[4].hist(df_2019['Score'], bins=20, color='mediumseagreen', edgecolor='black', alpha=0.7)
axes[4].set_title('2019')
axes[4].set_xlabel('Happiness Score')

# Comparación superpuesta
axes[5].hist(df_2015.iloc[:, 3], bins=20, alpha=0.5, label='2015', color='blue')
axes[5].hist(df_2016.iloc[:, 3], bins=20, alpha=0.5, label='2016', color='green')
axes[5].hist(df_2017.iloc[:, 2], bins=20, alpha=0.5, label='2017', color='red')
axes[5].hist(df_2018.iloc[:, 2], bins=20, alpha=0.5, label='2018', color='orange')
axes[5].hist(df_2019['Score'], bins=20, alpha=0.5, label='2019', color='purple')
axes[5].set_title('Comparación General')
axes[5].set_xlabel('Happiness Score')
axes[5].legend()

plt.tight_layout()
plt.show()
```

```python
# Tendencias de los países más felices
print("\nComparación de Top 5 Países Más Felices por Año:")
print("\n2015:")
print(df_2015.head(5)[['Country', df_2015.columns[3]]].to_string(index=False))
print("\n2016:")
print(df_2016.head(5).iloc[:, [0, 3]].to_string(index=False))
print("\n2017:")
print(df_2017.head(5).iloc[:, [0, 2]].to_string(index=False))
print("\n2018:")
print(df_2018.head(5).iloc[:, [1, 2]].to_string(index=False))
print("\n2019:")
print(df_2019.head(5)[['Country or region', 'Score']].to_string(index=False))
```

```python
# Estadísticas comparativas

stats_comparison = pd.DataFrame({
    '2015': df_2015.iloc[:, 3].describe(),
    '2016': df_2016.iloc[:, 3].describe(),
    '2017': df_2017.iloc[:, 2].describe(),
    '2018': df_2018.iloc[:, 2].describe(),
    '2019': df_2019['Score'].describe()
})

print("\n", stats_comparison.round(3))
```

```python
# Gráfico de líneas para mostrar tendencia del promedio
years = ['2015', '2016', '2017', '2018', '2019']
mean_scores = [
    df_2015.iloc[:, 3].mean(),
    df_2016.iloc[:, 3].mean(),
    df_2017.iloc[:, 2].mean(),
    df_2018.iloc[:, 2].mean(),
    df_2019['Score'].mean()
]

plt.figure(figsize=(12, 6))
plt.plot(years, mean_scores, marker='o', linewidth=2, markersize=10, color='darkblue')
plt.title('Tendencia del Promedio de Happiness Score (2015-2019)', fontsize=14, fontweight='bold')
plt.xlabel('Año', fontsize=12)
plt.ylabel('Happiness Score Promedio', fontsize=12)
plt.grid(True, alpha=0.3)
plt.ylim(min(mean_scores) - 0.1, max(mean_scores) + 0.1)

# Agregar valores en los puntos
for i, score in enumerate(mean_scores):
    plt.text(i, score + 0.02, f'{score:.3f}', ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```

## 📝 Conclusiones del EDA

Este análisis exploratorio de datos ha revelado:

1. **Distribución de Felicidad**: Los scores de felicidad siguen una distribución relativamente normal en todos los años.

2. **Países Líderes**: Los países nórdicos (Finlandia, Dinamarca, Noruega) consistentemente ocupan los primeros lugares.

3. **Factores Correlacionados**: Las variables económicas (GDP per capita) y sociales (Social support, Healthy life expectancy) muestran fuerte correlación con el happiness score.

4. **Tendencias Temporales**: El promedio de felicidad mundial ha mostrado variaciones mínimas entre 2015-2019.

5. **Valores Atípicos**: Existen algunos países con scores significativamente bajos o altos en comparación con la media.
